# Clean Data + Reorder Data

- Actually we have the data in different languages so we have to create a program to translate the different data

```python
import pandas as pd
# We will try to use a translator that aleady exist
from googletrans import Translator, constants
```

```python
translator = Translator()
# Await each time that translator function is done from spanish to english
```

```python
data = pd.read_excel("../data/1-Raw/Forms-Results-Raw.xlsx")
```

```python
data.head()
```

```python
# Translate Occupation
idx = 0
for occup in data['Occupation']:
    translation = await(translator.translate(occup, src="es", dest="en"))
    data['Occupation'][idx] = translation.text
    idx += 1

# Translate Country 
idx = 0
for country in data['Country']:
    translation = await(translator.translate(country, src="es", dest="en"))
    data['Country'][idx] = translation.text
    idx += 1
```

- Null data is only in Application question so its okay to not modify it
- We have to modify the data in a question do that its not a list and we need it as a list

```python
# Column What do you primarily use AI tools for? create as a tuple
idx = 0
for usages in data['What do you primarily use AI tools for?']:
    new_list = usages.split(',')
    data['What do you primarily use AI tools for?'][idx] = []
    for usage in new_list:
        data['What do you primarily use AI tools for?'][idx].append(usage)
    idx += 1
```

# Export as a other type of data

```python
data.to_csv('../data/2-Processed/Forms-Results-Raw.xlsx.csv', index=False)  
```
